```python
text = """"When I went to the cinema to see this film, my first thought was, "Wow, this looks really good, I'm sure I'll love it," but in the end, it didn't quite live up to my expectations. 
Don't get me wrong, I liked it and found it quite enjoyable, but there were just a few things I found unpleasant, and I wanted to point them out. 
To begin, let's talk about the good points: The animation is very good. 
I saw many people saying that it looked better than Spider-Verse (another Sony film, the studio that made this one), but I personally prefer the animation in Spider-Verse to this film's, although I still think it's outstandingly good in this aspect. 
On the other hand, I also have to point out that there were certain moments that were genuinely funny. 
There were some jokes that only made the children in the cinema laugh, but there were also many parts where I couldn't stop laughing. 
That pleasantly surprised me. 
Another important point is that the action in the matches is very good and frenetic; I really enjoyed watching the few matches that could be seen in detail, but at the same time I feel that this has worked against it a bit, because, as I've already mentioned, there are very few fights that can be seen in detail in this film, and it seems like more are needed. 
Something I liked at first, but which later deteriorated, is that the pacing was very dynamic; in fact, it was too much so. The film starts by jumping from one thing to another quite effectively. 
They begin a scene, finish it properly, and move on to something else, but at a certain point, they start overusing editing to speed things up, omitting scenes that could have been very interesting for the viewer (I mention again some matches that last only seconds on screen as an example of this) and proceed to resolve the interpersonal conflicts that arise in the story in a matter of seconds. 
I feel that, because of all this, the film could have benefited from an extra 10 minutes. In conclusion, Goat is a very fun and entertaining film, but in its quest to keep your brain releasing dopamine, it leaves out many things that would have been very welcome."""

print(text)
```

Extractive summary ---

```python
import nltk # natural language toolkit
from nltk import sent_tokenize
nltk.download('punkt_tab')

from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
import regex

newText = sent_tokenize(text) # each sentence is a token
newText
```

```python
cleanText = []

for sent in newText:
    sent = regex.sub(r"[^\w\s]", "", sent) # substitute *except* characters and spaces with ""
    sent = sent.lower()
    cleanText.append(sent)

cleanText
```

```python
tempText = []

for sent in cleanText:
    words = sent.split(' ')
    sent = []
    for word in words:
        if not word in stopwords.words('english'):
            sent.append(word)
    tempText.append(sent)

cleanText = tempText
cleanText # 2d list of sentences and words within
```

```python
from gensim.models import Word2Vec # word embeding -> words are converted to vectors, related words have similar vectors

embedder = Word2Vec(cleanText, vector_size=1, min_count=1, epochs=100)
```

```python
embedder.wv # embeder.wv is a list of vectors that correspond to words

vectors = []

for i in range(len(cleanText)):
    vectors.append([])
    for j in range(len(cleanText[i])):
        vectors[i].append(embedder.wv[cleanText[i][j]])

vectors
```

```python
longestLen = 0

for sent in cleanText:
    if len(sent) > longestLen:
        longestLen = len(sent)

longestLen
```

```python
lens = []

for sent in cleanText:
    lens.append(len(sent))
print(max(lens)) # another method of ^
```

```python
import numpy as np
# padding -> making all sentences same lenght as the longest

tempVectors = []

for sent in vectors:
    sent = np.pad(sent, (0, longestLen-len(sent))) # padding with 0s
    tempVectors.append(sent)

vectors = tempVectors
vectors
```
