```python
text = 'I will walk to the store'
print(text)

words = text.lower().split(' ')
words.sort()
print(words)

indexToWord = {}
index = 0
for word in words:
    indexToWord[index] = word
    index += 1
print(indexToWord)

wordToIndex = {}
index = 0
for word in words:
    wordToIndex[word] = index
    index += 1
print(wordToIndex)
```

```python
indexText = []
for word in text.lower().split(' '):
    indexText.append(wordToIndex[word])

indexText
```

```python
sequenceLength = 4

features = []
targets = []

for i in range(len(indexText)-sequenceLength):
    features.append(indexText[i:i+sequenceLength])
    targets.append(indexText[i+sequenceLength])

print(features)
print(targets)
```

```python
import numpy as np

features = np.array(features)
targets = np.array(targets)

print(features.shape)

featuresCat = np.zeros((features.shape[0], features.shape[1], len(words))) # features shape, categories
featuresCat
```

```python
lisIndex = 0

for lis in features:
    itemIndex = 0
    for item in lis:
        featuresCat[lisIndex][itemIndex][item] = 1
        itemIndex += 1
    lisIndex += 1

featuresCat
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN

model = Sequential([
    SimpleRNN(128, input_shape=(sequenceLength, len(words)), activation='relu'), # neurones, input shape (sequence length, amount of unique words)
    Dense(len(words), activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy']) # target is the actual value - not a class
model.fit(featuresCat, targets, epochs=50, batch_size=32, verbose=True)
```

```python
def preprocess(tex):
    texIndex = []

    newWords = tex.lower().split(' ')

    newWords = newWords[-sequenceLength:]

    for word in words:
        texIndex.append(wordToIndex[word])

    texCat = np.zeros((1, sequenceLength, len(words)))

    itemIndex = 0
    for i in range(sequenceLength):
        texCat[0][i][texIndex[i]] = 1

    return texCat

preprocess('I will walk to')
```

```python
prediction = model.predict(preprocess('I will'))
prediction = np.argmax(prediction)
prediction = indexToWord[prediction]
prediction
```
