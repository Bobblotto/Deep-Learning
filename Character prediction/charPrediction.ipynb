```python
text = 'I will walk to the store'
```

```python
chars = list(set(text)) # get all unique items
chars.sort()
chars
```

```python
charToIndex = {}

index = 0

for char in chars:
    charToIndex[char] = index
    index += 1

charToIndex

indexToChar = {}

index = 0

for i in range(len(chars)):
    indexToChar[i] = chars[i]

indexToChar
```

```python
indexText = []

for char in text:
    indexText.append(charToIndex[char])

indexText
```

```python
sequenceLength = 4

features = []
targets = []

index = 0

while index < len(indexText)-sequenceLength:
    sequenceList = indexText[index:index+sequenceLength]
    features.append(sequenceList)
    targets.append(indexText[index+sequenceLength])
    index += 1

print(features)
print(targets)
```

```python
import numpy as np

features = np.array(features)
targets = np.array(targets)
```

```python
ar = np.zeros((features.shape[0], features.shape[1], len(chars)))
ar
```

```python
lisIndex = 0

for lis in features:
    itemIndex = 0
    for item in lis:
        ar[lisIndex][itemIndex][item] = 1
        itemIndex += 1
    lisIndex += 1

featureCat = ar
featureCat
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN

model = Sequential([
    SimpleRNN(128, input_shape=(sequenceLength, len(chars)), activation='relu'), # neurones, input shape (sequence length, amount of unique characters)
    Dense(len(chars), activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy']) # target is the actual value - not a class
model.fit(featureCat, targets, epochs=50, batch_size=32, verbose=True)
```

```python
def preprocess(tex):


    if len(tex) < sequenceLength:
        tex = tex.zfill(sequenceLength) # fill to sequence length
    elif len(tex) > sequenceLength:
        tex = tex[-sequenceLength:] #truncate to sequence length

    texIndex = []

    for char in tex:
        texIndex.append(charToIndex[char])

    texCat = np.zeros((1, sequenceLength, len(chars)))

    itemIndex = 0

    for i in range(sequenceLength):
        texCat[0][i][texIndex[i]] = 1

    return texCat
```

```python
prediction = model.predict(preprocess('I walk t'))
prediction = np.argmax(prediction)
prediction = indexToChar[prediction]
prediction
```
