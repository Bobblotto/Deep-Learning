```python
text = 'I will walk to the store'
```

```python
chars = list(set(text)) # get all unique items
chars
```

```python
charToIndex = {}

index = 0

for char in chars:
    charToIndex[char] = index
    index += 1

charToIndex

indexToChar = {}

index = 0

for i in range(len(chars)):
    indexToChar[i] = chars[i]

indexToChar
```

```python
indexText = []

for char in text:
    indexText.append(charToIndex[char])

indexText
```

```python
sequenceLength = 4

features = []
targets = []

index = 0

while index < len(indexText)-sequenceLength:
    sequenceList = indexText[index:index+sequenceLength]
    features.append(sequenceList)
    targets.append(indexText[index+sequenceLength])
    index += 1

print(features)
print(targets)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN

model = Sequential([
    SimpleRNN(128, input_shape=(sequenceLength, len(chars)), activation='relu'), # neurones, input shape (sequence length, amount of unique characters)
    Dense(len(chars), activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy']) # target is the actual value - not a class
# model.fit
```
