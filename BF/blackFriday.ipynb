```python
import pandas as pd

train = pd.read_csv(r'C:\Users\eyomi\OneDrive\Pictures\VScode\Deep Learning\BF\train_black_friday.csv')
test = pd.read_csv(r'C:\Users\eyomi\OneDrive\Pictures\VScode\Deep Learning\BF\test_black_friday.csv')
train
```

```python
prodID = train['Product_ID']
prodID = prodID.astype('category').cat.codes # convert to integers (similar to label encoder)
train['Product_ID'] = prodID

prodID = test['Product_ID']
prodID = prodID.astype('category').cat.codes # convert to integers (similar to label encoder)
test['Product_ID'] = prodID

prodID.astype('category').cat.codes
```

```python
data = pd.concat([train, test])  # join 2 dataframes together naturally (concatonate)
data
```

```python
categorical = ['Gender', 'Age', 'City_Category', 'Stay_In_Current_City_Years']

for i in categorical:
    data[i] = data[i].astype('category').cat.codes # converting all categoricals to ints

data
```

```python
mean = data['Product_Category_2'].mean()
data['Product_Category_2'] = data['Product_Category_2'].fillna(mean)

mean = data['Product_Category_3'].mean()
data['Product_Category_3'] = data['Product_Category_3'].fillna(mean)

mean = data['Purchase'].mean()
data['Purchase'] = data['Purchase'].fillna(mean)

# removing all missing values

data.isna().sum()
```

```python
features = data.drop(columns=['Purchase', 'User_ID'])
target = data['Purchase']
features
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
features = scaler.fit_transform(features) # normalising
```

```python
from sklearn.model_selection import train_test_split

trainFeatures, testFeatures, trainTarget, testTarget = train_test_split(features, target, train_size=0.8)
```

```python
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor()
regressor.fit(trainFeatures, trainTarget)

prediction = regressor.predict(testFeatures)
prediction
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

error = mean_squared_error(prediction, testTarget)
error

mean_absolute_error(prediction, testTarget)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import Input

model = Sequential([
    Input(shape=(len(features), )),
    Dense(256, activation='relu'),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(1) # no classification
])

model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mean_squared_error'])
model.fit(trainFeatures, trainTarget, epochs=10, validation_data=(testFeatures, testTarget), verbose=1)
```
