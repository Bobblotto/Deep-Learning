```python
import numpy as np
import matplotlib.pyplot as plt

def cost(x):
    return x**2

def derivative(x): # getting the slope at a certain point
    return x*2

def gradDescent(x, epoch, learnRate):
    lis = [x]
    for i in range(epoch):
        slope = derivative(x)
        x = x-learnRate*slope
        lis.append(x)
    
    return np.array(lis)

x = 10
epoch = 5
learnRate = 0.1

des = gradDescent(x, epoch, learnRate)
print(des)
```

```python
xs = np.linspace(-x, x, 100)
y = cost(xs)
```

```python
plt.figure(figsize=(5, 5))
plt.plot(xs, y)
plt.scatter(des, cost(des))
plt.ylabel('Error')
plt.xlabel('Gradient Descent')
plt.show()
```
