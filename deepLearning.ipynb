```python
from sklearn.datasets import load_iris

data = load_iris()
print(data)

features = data['data']
target = data['target']
```

```python
from sklearn.preprocessing import OneHotEncoder

target = target.reshape(-1, 1) # make 2d

encoder = OneHotEncoder()

target = encoder.fit_transform(target)
print(target)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential([ # each item is a layer
    Dense(16, activation='relu', input_shape=[4, 1]), # amount of neurones(2^x), relu, 4-features, 1-target
    Dense(16, activation='relu'),
    Dense(3, activation='softmax') # 3 classifications
])
```
