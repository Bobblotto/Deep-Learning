```python
text = 'I will walk to the store'
print(text)

words = text.lower().split(' ')
words.sort()
print(words)

indexToWord = {}
index = 0
for word in words:
    indexToWord[index] = word
    index += 1
print(indexToWord)

wordToIndex = {}
index = 0
for word in words:
    wordToIndex[word] = index
    index += 1
print(wordToIndex)
```
